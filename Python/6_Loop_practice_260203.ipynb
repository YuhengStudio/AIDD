```python
# 列表计算

#IC50 列表
ic50_values = [0.05,0.12,0.08,0.15,0.03,0.02]

# 1.计算总和
total = 0
for ic50 in ic50_values:
    total += ic50
    #print(total)    这是不对的，因为每个循环都加一遍了，而这不需要，只需要一个结果
print(f"IC50总和为: {total:.2f} μM")

# 2.计算平均值这里注意，可以len，但这个是计算平均值的本质要掌握
count = 0
total = 0
for ic50 in ic50_values:
    total += ic50           #每一次循环 total多个ic50，count多1
    count += 1              #这个注意不是从0开始取，要加1
average = total / count
print(f"平均数为：{average:.2f} μM")


# 3.找出最小值（最好的化合物）  #假设 最小
min_ic50 = ic50_values[0]
for ic50 in ic50_values:
    if ic50 < min_ic50:
    #    print(ic50)       # 这种是不对的，因为不止一个值
      min_ic50 = ic50      # 应该循环赋值才对，另，注意这里要继续缩进 2 格
print(f"最小值为：{min_ic50} μM")
```

```python
# 筛选数据，使用 if

# IC50列表 
ic50_values = [0.23,0.05,0.12,0.08,0.15,0.02]

# 1.统计小于0.1的化合物数量
# 2.创建新列表，只包含小于0.1的值

count_potent = 0
potent_compounds = []

for value in ic50_values:
    if value < 0.1:
        count_potent += 1
        potent_compounds.append(value)
print(f"高活性化合物的数量：{count_potent}")
print(f"高活性IC50列表：{potent_compounds}")

```

```python
#分析SMILE字符串
smiles = "CC(=O)Oc1ccccc1C(=O)O"

# 统计脂肪碳数量，统计芳香碳数量，统计总碳数
upper_c_count = 0
lower_c_count = 0

for char in smiles:   #用 char代替 word
    if char == "C":
        upper_c_count += 1
    if char == "c":
        lower_c_count += 1
total_c_count = upper_c_count + lower_c_count

print(f"脂肪碳：{upper_c_count}个")
print(f"芳香碳：{lower_c_count}个")
print(f"总碳数：{total_c_count}个")

```

```python
#上面换一种思路，统计 total时候

smiles = "CC(=O)Oc1ccccc1C(=O)O"
upper_c_count = 0
lower_c_count = 0

total_c_count = 0

for char in smiles:
    if char == "C":
        upper_c_count += 1
        total_c_count += 1   #这样每次 total也增加一个
    if char == "c":
        lower_c_count += 1
        total_c_count += 1   #但这里 total是基于大写 C的结果继续增加，很巧妙


print(f"脂肪碳(C): {upper_c_count}")
print(f"芳香碳(c): {lower_c_count}")
print(f"总碳数: {total_c_count}")
```

```python
# 嵌套循环，生成九九乘法表

# 生成乘法表（1-9）
# 格式：
# 1 x 1 = 1
# 1 x 2 = 2
# ...
# 9 x 9 = 81

for i in range(1,10):        # 取 1-9
    for j in range(1,10):    # 取 1-9
        result = i * j
        print(f"{i} x {j} = {result}")
```
