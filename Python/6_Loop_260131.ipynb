````python
#使用循环
smiles_list = ["CCO","CC(=O)","c1ccccc1","CCN","CCC"]

for smiles in smiles_list:   # for 变量 in 序列:
    print(smiles)               # 循环体代码

# 要领如下：
# for-关键字，表示循环开始
# 变量- 循环变量（每次取序列中的一个元素）
# in- 关键字，表示（在....之中）
# 序列- 可遍历对象（列表，字符串都行） 
# 冒号- 表示循环体开始 （缩进 4个空格）

````

````python
#遍历列表

# 示例 1：遍历药物名称
drugs = ["Aspirin","Ibuprofen","Pararcetamol"] # drugs是列表变量
for drug in drugs:
    print(drug)

#
#**执行过程详解：**
#```
#第1次循环：drug = "Aspirin"    → 执行print("Aspirin")
#第2次循环：drug = "Ibuprofen"  → 执行print("Ibuprofen")
#第3次循环：drug = "Paracetamol"→ 执行print("Paracetamol")

#循环结束
````

````python
#循环变量名字：可以任意取，但要有意义

for drug in drugs:
    print(drug)    #单数对应复数，很好！ #并且drug是变量，不用加引号

for d in drugs:
    print(d)      #没毛病，但不够清晰！
for compound in drugs:
    print(compound)   #没毛病

for x in drugs:
    print(x)    #不好，看不出是啥，我们要起一个好辨认的 “循环变量”名字
    
````

````python
# 循环中，缩进的重要性
# 循环体的所有缩进都会重复执行

drugs = ["Aspirin", "Ibuprofen", "Paracetamol"]  # 顶层代码必须顶格！！

for drug in drugs:
    print(f"当前药物：{drug}")  # 非顶层代码，可以任意缩进（几个要平行就行）
    length = len(drug)         # print length平行，但一般缩进 4格
    print(f"名称长度：{length}")
    print("-"*20)



# 发现每次循环，都是从上到下执行缩进的全部行。。。
# 但如果把 print改变缩进的位置，比如缩进3个格，那就报错
    
````

````python
#计算多个化合物分子量, 注意total初始 0

molecular_weights = [180.16,206.28,151.16,308.33]
print("分子量列表：")
for mw in molecular_weights:
    print(f"{mw}g/mol")



#计算总和

total = 0
for mw in molecular_weights:
    total = total + mw 
print(f"总分子量为：{total:.2f}g/mol")

#计算平均值   #注意，count不是自动计数
count = 0
total = 0
for mw in molecular_weights:
    count = count + 1     # （和序数递增无关）0开始，第二轮变成 1 , 1再带入关系变成 2
    total = total + mw    # 第一个的结果累加，第二轮继续
average = total/count     # 这步不用循环了，毕竟只要两个结果进行一次运算

print(f"平均分子量：{average:.2f}g/mol")
````

````python
# 用range遍历列表
drugs = ["Aspirin","Ibuprofen","Paracetamol"]

#len(drugs) =3 , range(3) = 0,1,2 
for i in range(len(drugs)):        # 这样设计range循环，为了012...
    print(f"索引{i}: {drugs[i]}")  # 前面遍历012，后面遍历每个元素
                                  # 因为 len就可以算出drugs[i]需要的 i
    
````

````python
# 同时处理两个列表

compound = ["Aspirin","Ibuprofen","Paracetamol"]
mw_values = [180.16,206.28,151.16]

for i in range(len(compound)):
    pound = compound[i]
    mw = mw_values[i]
    print(f"{pound}的分子量为: {mw}")
````

````python
# 更优雅的 enumerate(列举) ，返回值是（索引，值） ，索引是0,1,2....

#类似于
#(0, "Aspirin")
#(1, "Ibuprofen")
#(2, "Paracetamol")

drugs = ["Aspirin","Ibuprofen","Paracetamol"]

for index,drug in enumerate(drugs):  # index随便用别的取代，比如 i(不可用数字）
    print(f"索引：{index},{drug}")
````

````python
# 如果从 1 开始编号呢？ 直接 enumerate （，start=1 ）

compounds = ["Aspirin","Ibuprofen","Paracetamol"]
ic50_values = [0.05,0.12,0.08]

# 输出：
# 化合物活性报告
# ========================================
# 1. Aspirin: IC50 = 0.05 μM
# 2. Ibuprofen: IC50 = 0.12 μM
# 3. Paracetamol: IC50 = 0.08 μM

print(f"化合物活性报告")
print(f"="*40)
for index,compound in enumerate(compounds,start=1): # 编号从0改为1开始
    ic50 = ic50_values[index-1] #刚才index
    print(f"{index}.{compound}: IC50 = {ic50:.2f} μM")
    

            
````

````python
# AIDD 统计SMILES中原子个数
smiles = "CC(=O)Oc1ccccc1C(=O)O"

c_count = 0   # 预设一个碳计数 0开始
o_count = 0   # 。。.氧 0开始

for char in smiles:
    if char == "C" or char == "c":    # 注意这是（双）代表等号，不是赋值
        c_count += 1   
    if char == "O":
        o_count += 1
print(f"碳原子数：{c_count}")
print(f"氧原子数:{o_count}")

````

````python
#嵌套循环  
#循环里面套循环

#外层循环
for i in range(3):   #循环的是序号
    print(f"外层循环{i}次")  # 外循环先取 0
    for j in range(2):      # 基于外循环 0 ，再 内循环 0.1.。。。
        print(f"  内层循环{j}次")
````

````python
# AIDD 练习一个嵌套循环    # 嵌套为了

drugs = ["Aspirin","Ibuprofen","Paracetamol"]

print(f"所有药物配对:")
for drug1 in drugs:        # for 缩进
    for drug2 in drugs:    # 嵌套的目的只是基于第一行a ，选出a b c，组成aa ab ac
        
        if drug1 != drug2:  # if 也要缩进，另 !=是和 ==相似，不是 =（赋值）相似
            print(f"{drug1} vs {drug2}")

#分析如下 ：

#第一行取 A

#第二行开始
# A  
# B  
# C  

#第三行开始

# A = A , 不输出
# B !=A , 输出
# C !=A , 输出

#第四行开始
# 输出 A vs B  A vs C   B vs A    B vs C   C vs A   C vs B 

#再重复第一行取B......


````
