```python
#连接，把列表变成字符串（列表中真实的逗号删掉，加入字符虚假的逗号或者其他的任何字符都行）
drug_list = ['Aspirin', 'Ibuprofen', 'Paraceamol']
result = ",".join(drug_list)
print(result)

#换个逗号为句号 
#甭管啥号，都是字符而已
result2 = "。".join(drug_list)
print(result2)

#处理aidd的csv数据
csv_line = "Aspirin,180.16,1.23" #用切割split去识别逗号这个字符串
fields = csv_line.split(",")
print(fields)
name = fields[0]
mw = float(fields[1])
logp = float(fields[2])
print(f"{name}:MW={mw},LogP={logp}")
```

```python

#旧式格式化(%操作符）
drug = "Aspirin"
mw = 180.16 
logp = 2 
message = "%s的分子量是%.2f,logp是%d" %(drug,mw,logp)
print(message) 
```

```python
#format()方法格式化，格式化指的是改变格式
drug = "Aspirin"
mw = 180.16443 
logp = 2 
message = "{}的分子量是{:.3f}".format(drug,mw)  
#第一个括号不用调整格式化，不用加冒号，第二个有指令需要冒号
print(message)
```

```python
#指定位置，0和1替换
drug = "Aspirin"
mw = 180.16

message = "{}的分子量是{}".format(drug, mw) #按顺序
print(message)

#当后面括号顺序颠倒，可以用1和0代替
message = "{1}的分子量是{0}".format(mw,drug)
print(message)  # Aspirin的分子量是180.16
```

```python
#f-string
drug = "Aspirin"
mw = 180.166
logp = 1.23333

#直接在字符串中嵌入变量
message = f"{drug}的分子量是{mw}"
print(message)

#格式控制保留小数
message = f"{drug}:MW={mw:.2f},LogP={logp:.3f}"
print(message)

#可以在{}中写表达式
num = 10
print(f"10的平方是{num ** 2}")


#AIDD应用  #有效数字（科学计数法.2e）和直接普通小数.2f
compound = "Aspirin"
ic50 = 0.000000325
print(f"{compound}的IC50值是{ic50:.2e}M") #科学计数法

print(f"{compound}的IC50值是{ic50:.2f}M") #小于0.01就是0.00
```
