```python
# Aspirin的SMILES   新知识：count 和 in
smiles = "CC(=O)Oc1ccccc1C(=O)O"

#1计算长度 
length = len(smiles)
print(f"smiles的长度：{length}")

#2第一个c的位置，用find查看位置（找不到就是 -1）
position = smiles.find("c")
print(f"第一个c出现位置:{position}")


###3检查出现“c” ,用in运算符输出--布尔值
has_aromatic = "c" in smiles
print(f"是否出现c：{has_aromatic}")

##统计“C”出现次数，用count
c_count = smiles.count("c")
print(f"出现几次c:{c_count}次")

#提取前十个字符
prefix = smiles[:10]
print(f"前十个字符：{prefix}")
```

```python
#混淆 in 和 find()  而-1也是在if中当做True
smiles = "CC(=O)O"

print("(=O)" in smiles)

print(smiles.find("(=O)"))

if smiles.find("XYZ"): 
    print("找到了")   # 找不到返回-1，但-1也是True！这是不对的

#需要显式判断！

if smiles.find("XYZ") != -1:
    print("找到了")
else:
    print("没找到")

#另外用in更简单

if "XYZ" in smiles:
    print("找到了")
else:
    print("没有")
```

```python
#作业1：SMILES分析
#给定SMILES："c1ccccc1C(=O)O"

#计算长度 #检查是否包含羧基 "C(=O)O" #提取前6个字符（苯环部分） #统计小写c的数量（芳香碳原子）
smiles = "c1ccccc1C(=O)O"
length = len(smiles)
has_cooh = "C(=O)O" in smiles
arom_six = smiles[:6]
c_count = smiles.count("c")

print(f"长度：{length}\n有没有羧基：{has_cooh}\n前六个字符：{arom_six}\n小写c数量：{c_count}")
```
