```python
#作业2：板孔计算,实验室有150个样品，使用96孔板：

#需要多少个完整的板？
#最后一个板用了多少孔？
#最后一个板还剩多少空孔？


#我的作业
import math
sample = 150
plate_count = math.ceil(sample / 96)
well_last = sample % 96

well_remaining = 96 - well_last

print(plate_count,well_last,well_remaining)


#标准答案：
import math

samples = 150
wells_per_plate = 96

# 板子数量（向上取整）
plate_count = math.ceil(samples / wells_per_plate)

# 最后一板使用的孔数
well_last = samples % wells_per_plate

# 如果刚好用满一板，余数会是 0，需要特殊处理
if well_last == 0:
    well_last = wells_per_plate

# 最后一板剩余空孔
well_remaining = wells_per_plate - well_last

print(f"需要的板数: {plate_count}")
print(f"最后一板使用的孔数: {well_last}")
print(f"最后一板剩余的空孔: {well_remaining}")


```

```python
# 作业3：LogP计算，
# 化合物在正辛醇和水中的浓度分别是 8.5e-4 M 和 3.2e-6 M，计算LogP值。

octanol_concentration = 8.5e-4 
water_concentration = 3.2e-6
import math
logp = math.log10(octanol_concentration / water_concentration)
print("LogP:" ,logp)

#用f-string保留小数

print(f"LogP:{logp:.2f}")

```
