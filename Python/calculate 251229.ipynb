```python
# 使用数学函数，否则容易出错
# 问题：100个样品，每板96孔，问需要多少个板？
# 问题：有多少样品在最后一板？

samples = 100
wells_per_plate = 96

# 误区：如果直接 
#plate = samples // wells_per_plate + 1 
#print(plate) # 输出也是2，看起来对，但是如果样品刚好96个，那结果也是2，但是实际是1个板

# 则需要用到四舍五入（向上取整）
import math
plate = math.ceil(samples / wells_per_plate)
sample_last = samples % wells_per_plate
print(f"{plate},{sample_last}") # 使用f-string和{}更舒适 


#同理还有math.floor（）向下取整
```

```python
# 96孔板，每行12个孔，注意计算机的行列都是从0开始便于统计（0-11）
well_number = 25
row = well_number // 12  # 第几行（从0开始）
column = well_number % 12  # 第几列（从0开始） column是列的意思
print(f"第{row}行，第{column}列")  # 第2行，第1列  

well_number = 12
row = well_number // 12 
column = well_number % 12
print(f"第{row}行，第{column}列")
```

```python
# AIDD应用：四舍五入 round是个内置函数
molecular_weight = 180.16342
rounded = round(molecular_weight, 2)  # 保留2位小数
print(rounded)  # 180.16
```
