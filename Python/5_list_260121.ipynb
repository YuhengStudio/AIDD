```python
# 虚拟筛选结果（按得分排序）
compounds = ["Compound_A", "Compound_B", "Compound_C", "Compound_D", "Compound_E"]
scores = [9.5, 8.7, 8.3, 7.9, 7.2]

# 选择Top 3
top4_compounds = compounds[:4] #在终点4号之前就停下，不包括4号
top4_scores = scores[:4]

print("Top4化合物:")
for i in range(4):  #for xx in代表循环，这俩绑定在一起
    #冒号代表循环体，第二行要缩进，缩进属于for循环内部的代码，只执行这一行
    
    print(f"{top4_compounds[i]}: {top4_scores[i]}") #i可以随便替换，代表index而已，但range不可以换，是内置函数
```

```python
#指定位置插入 insert() #只能增加选定位置的一个元素！！

drugs = ["Aspirin", "Paracetamol"]
drugs.insert(1,"Ibuprofen")   #insert插入的是字符串，进入列表就是元素
print(drugs)

drugs.insert(0,"Warfarin")
print(drugs)
```

```python
#添加另一个列表所有元素 ，用extend()

drugs1 = ["Aspirin", "Ibuprofen"]
drugs2 = ["Paracetamol", "Naproxen"]
drugs1.extend(drugs2)

print(drugs1)  #在1列表中添加2,括号里是新添加的，元素列在1后面。
print(drugs2) #列表2不变，因为主体是列表1


#列表之间用 “+”,也相当于上面的extend
drugs3 = ["Aspirin"] + ["Ibuprofen"] + ["Paracetamol"]
print(drugs3)
```

```python
# append & extend 对比
#字面上 附属 & 延伸

list1 = [1,2,3]
list1.append([4,5]) #虽然不用加引号，
                    #但输出的是把包括方括号在内都当成了一个单独的元素，加进去了
print(list1)  

#或者用普通字符串
list0 = [1,2,3]
list0.append("4,5") #把字符串加进去了，要引号，但这也不是正常的5个元素的列表
print(list0)  

#只有这样才是列表扩展！
list2 = [1,2,3]
list2.extend([4,5]) #这是拆开来加，加多个元素
print(list2)   
```

```python
#删除元素 remove 和 append是对应的

drugs1 = ["Aspirin", "Ibuprofen", "Aspirin", "Paracetamol"] #列表重复了阿司匹林
drugs1.remove("Aspirin")
print(drugs1)
 
#相反， pop()可以删除指定元素  #并且pop会保留删除后的值为第二次赋值
drugs2 = ["Aspirin", "Ibuprofen", "Paracetamol"]
last = drugs2.pop()  #空括号，默认删除最后的元素

print(last)  #  #last把最后一个从drugs里面删了，但呆在last这里了
print(drugs2)

first = drugs2.pop(0)
print(first) #删除部分输出
print(drugs2) #基于上一步删除之后继续删除
```

```python
# 用sort(类别）排序整理，按照从小到大

ic50_values = [0.12, 0.05, 0.08, 0.03]
ic50_values.sort()             #数字从小到大,已经修改了"原列表"
print(ic50_values)             #输出原列表，也是修改了


# 倒序排序
ic50_values.sort(reverse=True) # 参数名=参数值
print(ic50_values) 


# 字符串排序（按字母顺序）
drugs = ["Paracetamol", "Aspirin", "Ibuprofen"] #字符串--首字母排序
drugs.sort()
print(drugs)

drugs2 = "Paracetamol"
drugs.sort()
#print(drugs2)   sort不会排列字符串,要用sorted

```

```python
#sorted()不改变原列表，ed相比sort就是修改好了

original = [3,1,4,1,5]
sorted_list = sorted(original)  #虽然赋新的值，但也不改原来列表

print(original)
print(sorted_list)

#还可以排序字符串
print(sorted("Aspirin"))
```

```python
#reverse() ，原地修改列表自身
drugs = ["Ispirin", "Abuprofen", "Paracetamol"]
drugs.reverse()  #不看首字母，只看列表元素整体的位置
print(drugs)  # drugs已经变了

#或者用切片
#并且之前drugs已经倒序了原列表，变成了P A I顺序
reversed_drugs = drugs[::-1] #相当于[none:none:-1]  [start:end:step]

print(reversed_drugs)
```

```python
#列表复制，注意：直接赋值不是复制列表

list1 = [1,2,3]
list2 = list1  #指向同一个列表了

#那么
list2.append(4)
print(list1)
print(list2) #两个都一样了，所以这不是复制


```

```python
#正确的复制方法：用[:]

#1 切片
list1 = [1,2,3]
list2 = list1[:]  #这个相当于[start:end:step] 也就是[0:len(list1):1]
list2.append(4)
print(list1) #和原来一样
print(list2) #重新复制一份，再修改


#2 copy()方法
list11 = [1,2,3]
list22 = list1.copy()
list22.append(4)
print(list11) #和原来一样
print(list22) #重新复制一份，再修改

#3 list函数
list111 = [1,2,3]
list222 = list(list111)  #list函数可以复制列表
list222.append(5)
print(list111)  #不变
print(list222)  
```
