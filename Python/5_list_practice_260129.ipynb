```python
#内置函数和方法的区别

#sorted_list = sorted(original)       内置函数 built-in functions

#ic50_values.sort()   #drugs.reverse()   方法 methods



# 分子量列表（未排序）
molecular_weights = [206.28, 180.16, 151.16, 308.33, 244.27]
print(f"原列表：{molecular_weights}")

# 1. 从小到大排序（不要修改原列表）  #
sorted_mw = sorted(molecular_weights)
print(f"升序：{sorted_mw}")

# 2. 从大到小排序（不要修改原列表）
reversed_mw = sorted_mw[::-1]
print(f"降序：{reversed_mw}")

sorted_mw.reverse() #列表的方法，修改了原列表
print(f"翻转升序列表：{sorted_mw}")


# 3. 从大到小排序 （内置函数，不修改原列表）
reversed_mw = sorted_mw[::-1]
print(f"翻转降序：{reversed_mw}")
```

```python
#SMILES处理  #学会用 max

smiles_list = [
    "CC(=0)0c1ccccc1C(=0)0",           #Aspirin
    "CC(C)Cc1ccc(cc1)C(C)C(=0)0",       #Ibuprofen
    "CC(=0)Nc1ccc(0)cc1"               #Paracetamol
]

#1.计算每个SMILES长度，存入新列表

smiles_lengths = []
for smiles in smiles_list:    #注意循环的是smiles，
    smiles_lengths.append(len(smiles))#从左到右循环字符串而已，不是0123(range才是索引)
print(f"SMILES长度：{smiles_lengths}")

#2.找出最长的SMILES

max_length = max(smiles_lengths)
print(f"最长的SMILES长度: {max_length}")


#3.检查是否所有SMILES都包含"C"  # 这个超纲，先领会下。。。

#答案：

all_contain_c = all("C" in smiles for smiles in smiles_list)

# "C" in smiles 这是布尔值判断  

#生成式语法顺序python看到的先是 "c" in smiles(你要生成什么）
#再看到 for smiles in xxx （你要从哪里取数据）

#all() 函数 只要有一个False就返回 False

print(f"都包含C: {all_contain_c}")  
    
    
```

```python
#列表复制，不等于直接复制

#!.比如这样，给两个都改变了
list1 = [1,2,3]
list2 = list1

list2.append(4)
print(list1)   #虽然通过2增加，把1也改变了

#正确做法

list1_new = [1,2,3]
list2_new = list1_new.copy()  #这是一个方法 method，不改变list1

list2_new.append(5)
print(list1_new)     #不改变
print(list2_new)
```
