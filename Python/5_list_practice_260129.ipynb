```python
#列表练习

# 创建一个包含5个药物名称的列表
drugs = ["Aspirin", "Ibuprofen", "Paracetamol", "Naproxen", "Diclofenac"]

# 1. 输出列表长度
print(f"列表长度为:{len(drugs)}")


# 2. 输出第一个药物
print(f"第一个药物为：{drugs[0]}")


# 3. 输出最后一个药物
print(f"最后一个药物：{drugs[-1]}")

# 4. 输出前三个药物
print(f"前三个药物为：{drugs[:3]}")

# 5.输出最后两个药物
print(f"最后两个药物：{drugs[-2:]}")
```

```python
# 5个化合物的IC50值（μM）  #新知识点：min/max/sum。。。。

ic50_values = [0.12, 0.05, 0.18, 0.03, 0.09]

# 1. 找出最小IC50（最佳化合物） #用min()
best_ic50 = min(ic50_values)
print(f"最佳IC50：{best_ic50}")

# 2. 找出最大IC50（最差化合物） #用max()
worst_ic50 = max(ic50_values)
print(f"最差IC50：{worst_ic50}")

# 3. 计算平均IC50 
average_ic50 = sum(ic50_values) / len(ic50_values)
print(f"平均IC50：{average_ic50}")



# 4. 统计IC50小于0.1的化合物数量（提示：需要循环，下节课会学）
# 暂时可以手动数

#先
```

```python
#内置函数和方法的区别

#sorted_list = sorted(original)       内置函数 built-in functions

#ic50_values.sort()   #drugs.reverse()   方法 methods



# 分子量列表（未排序）
molecular_weights = [206.28, 180.16, 151.16, 308.33, 244.27]
print(f"原列表：{molecular_weights}")

# 1. 从小到大排序（不要修改原列表）  #
sorted_mw = sorted(molecular_weights)
print(f"升序：{sorted_mw}")

# 2. 从大到小排序（不要修改原列表）
reversed_mw = sorted_mw[::-1]
print(f"降序：{reversed_mw}")

sorted_mw.reverse() #列表的方法，修改了原列表
print(f"翻转升序列表：{sorted_mw}")


# 3. 从大到小排序 （内置函数，不修改原列表）
reversed_mw = sorted_mw[::-1]
print(f"翻转降序：{reversed_mw}")
```
